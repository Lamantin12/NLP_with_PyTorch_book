#  Coding of language data

One-hot encoding

```python
from sklearn.feature_extraction.text import CountVectorizer
import seaborn as sns

corpus = [
    'Time flies like an arrow.',
    'Fruit flies like a banana.'
]
vocab = [
    'an',
    'arrow',
    'banana',
    'flies',
    'fruit',
    'like',
    'time'
]
one_hot_vectorizer = CountVectorizer(binary=True)
one_hot = one_hot_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(one_hot, annot=True, cbar=False, xticklabels=vocab, yticklabels=["Предложение1", "Предложение2"])
```

TF-IDF encoding

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vocab = [
    'an',
    'arrow',
    'banana',
    'flies',
    'fruit',
    'like',
    'time'
]
tfidf_vectorizer = TfidfVectorizer()
tridf = tfidf_vectorizer.fit_transform(corpus).toarray()
sns.heatmap(tridf, annot=True, cbar=False, xticklabels=vocab, yticklabels=["Предложение1", "Предложение2"])
```
